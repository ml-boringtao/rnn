```python
import os
import os.path
import cv2
import numpy as np
```

```python
root = 'datasets/UCF-101/'
include = ['Basketball']
```

```python
full_categories = [x[0] for x in os.walk(root) if x[0]][1:]
```

```python
categories = [c for c in full_categories if c in [os.path.join(root, e) for e in include]]
#categories = [c for c in categories if c not in [os.path.join(root, e) for e in exclude]]
```

```python
categories
```

```python
data = []
for c, category in enumerate(full_categories):
    files = [[os.path.join(dp, f), os.path.splitext(f)[0]] 
             for dp, dn, filenames in os.walk(category) 
             for f in filenames if os.path.splitext(f)[1].lower() in ['.avi','.flv','.mpg']
            ]
    for file in files:
        data.append({'path':file[0], 'name':file[1], 'label':category.split('/')[2]})
```

```python
len(data)
```

```python
def extractFrames(videos, frame_path):
    for video in videos:
        file_path = frame_path + video['label'] + '/'
        directory = os.path.dirname(file_path)

        try:
            os.stat(directory)
        except:
            os.mkdir(directory) 
        
        cap = cv2.VideoCapture(video['path'])
        current_frame = 0
        frame_id = 0
        total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        spread = int(np.floor(total / 9))
        while(cap.isOpened()):
            if current_frame < total:
                cap.set(1, current_frame)
                ret, frame = cap.read()
                if ret == True:
                    frame = cv2.resize(frame, (80, 60))
                    name = file_path + video['name'] + '_' + str(frame_id) + '.jpg'
                    print ('Creating...' + name)
                    cv2.imwrite(name, frame)
                    current_frame += spread 
                    frame_id += 1
                else:
                    break
            else:
                break
        cap.release()
    cv2.destroyAllWindows()
```

```python
extractFrames(data, 'datasets/UCF-101-frames-resize/')
```
